```python
#read csv and check columns
import pandas as pd
df = pd.read_csv('Unicorn_Companies.csv')
df.head()
df.columns
```

```python
# Step 3 - Connect to SQLite Database

import sqlite3
# Create database
conn = sqlite3.connect('unicorns.db')
cursor = conn.cursor()
```

```python
# Step 4 - Create table with correct columns

cursor.execute('''
CREATE TABLE unicorns (
    Company TEXT,
    Valuation_B REAL,
    Date_Joined TEXT,
    Country TEXT,
    City TEXT,
    Industry TEXT,
    Select_Investors TEXT,
    Founded_Year INTEGER,
    Total_Raised TEXT,
    Financial_Stage TEXT,
    Investors_Count INTEGER,
    Deal_Terms TEXT,
    Portfolio_Exits INTEGER
)
''')
conn.commit()
```

```python
# step 5 - Insert data into the table
df.rename(columns={
    'Valuation ($B)': 'Valuation_B',
    'Date Joined': 'Date_Joined',
    'Select Inverstors': 'Select_Investors',
    'Founded Year': 'Founded_Year',
    'Total Raised': 'Total_Raised',
    'Financial Stage': 'Financial_Stage',
    'Investors Count': 'Investors_Count',
    'Deal Terms': 'Deal_Terms',
    'Portfolio Exits': 'Portfolio_Exits'
}, inplace = True)

# Now insert into sql

df.to_sql('unicorns', conn, if_exists='append', index = False)
```

Now , RUN different queries to see required data from the dataset

```python
#step 6 - RUN QUERIES!
#Top 5 Countries with Most Unicorns

query = '''
SELECT Country, COUNT(*) as Total_Unicorns
FROM unicorns
GROUP BY Country
ORDER BY Total_Unicorns DESC
LIMIT 5
'''
pd.read_sql_query(query, conn)
```

```python
#Top Industries

query = '''
SELECT Industry, COUNT(*) as Total_Companies
FROM unicorns
GROUP BY Industry
ORDER BY Total_Companies DESC
LIMIT 5
'''
pd.read_sql_query(query, conn)
```

```python
# Oldest Unicorns
query = '''
SELECT Company, Founded_Year
FROM unicorns
ORDER BY Founded_Year ASC
LIMIT 5
'''
pd.read_sql_query(query, conn)
```
